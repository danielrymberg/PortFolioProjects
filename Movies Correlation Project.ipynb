```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Now we need to read in the data
df = pd.read_csv("movies.csv")
```

```python
# Now let's take a look at the data

df
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

df.dtypes
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
#Drop any duplicates

df.drop_duplicates()
```

```python
# Order our Data a little bit to see

df = df.sort_values(by=["gross"], inplace=False, ascending=False)
```

```python
df
```

```python
#Scatter plot with budget vs gross

plt.scatter(x=df["budget"], y=df["gross"], alpha=0.5)

plt.title("Budget vs Gross Earning")

plt.xlabel("Gross Earnings")

plt.ylabel("Budget for Film")

plt.show
```

```python
df.head()
```

```python
#Plot the budget vs gross using seaborn

sns.regplot(x="gross", y="budget", data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
#Plot the score vs gross using seaborn

sns.regplot(x="score", y="gross", data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Correlation Matrix between all numeric columns

df.corr(method ='pearson')
```

```python
df.corr(method ="spearman")
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr(method="pearson")

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation Matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized = df

for col_name in df.numerized.columns:
    if(df.numerized[col_name].dtype == "object"):
        df.numerized[col_name] = df.numerized[col_name].astype("category")
        df.numerized[col_name] = df.numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method="pearson")

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation Matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
pd.set_option("display.max_rows", None)
```

```python
sorted_pairs
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```

```python
df = pd.read_csv("movies.csv")
```

```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
# Looking at the top 15 compaies by gross revenue per year
```

```python
CompanyGrossSum = df.groupby(['company', 'year'])[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company','year'], ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
